```python
import torch
import torch.nn as nn

def create_embedding_layer(vocab_size, d_model):
    return nn.Embedding(vocab_size, d_model)

def embed_tokens(embedding_layer, tokens):
    return embedding_layer(tokens)

vocab_size = 20
d_model = 8

embedding = create_embedding_layer(vocab_size, d_model)

tokens = torch.tensor([2, 4, 5, 6, 3])

embedded = embed_tokens(embedding, tokens)

print("Token IDs:", tokens)
print("\nEmbedding shape:", embedded.shape)
print("\nEmbeddings:\n", embedded)
```
